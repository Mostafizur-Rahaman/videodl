```python
import yt_dlp
```

```python
urls = ["https://www.youtube.com/watch?v=fJwWzYXfpgw", "https://www.youtube.com/shorts/0jOMewDuqMI", "https://www.facebook.com/noumanbayyinah/videos/932801798247735", "https://www.facebook.com/reel/711756527795453"]



ydl_opts = {
    'outtmpl': 'downloads/%(title).80s.%(ext)s',                # Output file naming (organized)
    'format': 'bestvideo+bestaudio/best',                       # Best video + audio combination (important for YouTube)
    'merge_output_format': 'mp4',                               # Merge video & audio into mp4
    'noplaylist': False,                                        # Allow playlists or single videos
    'quiet': False,                                             # Quiet mode (set False if you want logs)
    'writesubtitles': False,                                    # Download subtitles (if available)  
    'subtitleslangs': ['en'],
    'writeautomaticsub': False,
    'writethumbnail': False,                                    # Thumbnail download
    'retries': 5,                                               # Retry handling   
    'fragment_retries': 5,
    'ignoreerrors': True,                                       # Ignore errors (useful for playlists)
    'restrictfilenames': True,                                  # Limit filename issues
    'overwrites': False,                                        # Overwrite existing files
    }
for url in urls:
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        try:
            ydl.download([url])
        except:
            pass
```
